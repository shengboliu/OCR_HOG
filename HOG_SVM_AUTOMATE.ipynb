```python
import cv2
import matplotlib.pyplot as plt
from skimage.feature import hog
from skimage import data, color, exposure
import numpy as np
import os
```

```python
def getHOGDescriptor(imagePath):
    im = cv2.imread(imagePath)
    gr = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY) 
    gr1 = cv2.resize(gr,(64*2,64*2))

    fd, hog_image = hog(gr1, orientations=9, pixels_per_cell=(4, 4),
                    cells_per_block=(1, 1), visualise=True)
    return fd
```

```python
target = [] 
HOG_DESC = []
path = '/home/spratik/OCR_HOG/positive_image'
for root ,dir , file in os.walk(path):
    #print root
    dirName = root.split('/')
    count = 0
    for name in file:
       #print name
       imagePath = os.path.join(root,name)
       dec = getHOGDescriptor(imagePath)
       HOG_DESC.append(dec)
       #print imagePath
       count += 1
    num = dirName.pop() 
    #print "dirName: " + num + "--> count:" + str(count)
    for n in range(count): 
        target.append(int(num)) 
        
print target     
print HOG_DESC        
```

```python
hog_des = np.array(HOG_DESC)
print hog_des
```

```python
print hog_des.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(hog_des, target,test_size = 0.30,random_state=0)
```
